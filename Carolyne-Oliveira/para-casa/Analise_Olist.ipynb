### 3. O tempo limite de entrega das compras está sendo respeitado?

```python
# Porcentagem de pedidos com base na classificacao
df_vendas_olist["status_entrega"].value_counts(normalize=True)*100
```

- Cerca de 72,76% dos pedidos sao entregues com atraso e cerca de 18.32% sao entregues adiantados, e somente 8,9% dos pedidos sao entregues dentro do prazo.

## **Visualização**

### - Média do Valor de Compras por Cliente e do Frete

```python
# Calculando as médias
media_preco_compra = df_vendas_olist['price'].mean()
media_preco_frete = df_vendas_olist['freight_value'].mean()

# Arredondando para 2 casas decimais
media_preco_compra_arredondada = round(media_preco_compra, 2)
media_preco_frete_arredondada = round(media_preco_frete, 2)

# Criando os dados para o gráfico
labels = ['Preço Médio das Compras', 'Preço Médio do Frete']
valores = [media_preco_compra_arredondada, media_preco_frete_arredondada]

# Configurando o gráfico de barras
plt.figure(figsize=(8, 5))
sns.barplot(x=labels, y=valores, hue=labels, palette='pastel', legend=False)

# Adicionando rótulos de valor nas barras
for i, v in enumerate(valores):
    plt.text(i, v + 1, f'{v:.2f}', ha='center', va='bottom', fontsize=12)

# Título e rótulos
plt.title('Comparação entre Preço Médio das Compras e do Frete', fontsize=16)
plt.ylabel('Valor em R$', fontsize=12)
plt.xlabel('Categorias', fontsize=12)

# Exibindo o gráfico
plt.show()
```

A **média de preço das compras** é de R$ 119,98, o que indica que, em média, cada cliente gasta cerca de R$ 120,00 em suas compras na plataforma. Já a **média de preço do frete** é de R$ 19,95, mostrando que o custo médio de envio por pedido é aproximadamente R$ 20,00. Esses valores ajudam a entender o comportamento de compra dos clientes e o impacto dos custos de frete no valor final.
