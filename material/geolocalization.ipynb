```python
import folium
from folium.plugins import HeatMap

# Lista de pontos (latitude, longitude)
# data = [
#     [-23.550520, -46.633308],
#     [-23.551520, -46.632308],
#     [-23.552520, -46.631308],
#     [-23.553520, -46.630308],
#     [-23.554520, -46.629308],
#     [-23.555520, -46.628308],
#     [-23.556520, -46.627308],
#     [-23.557520, -46.626308],
#     [-23.558520, -46.625308],
#     [-23.559520, -46.624308],
# ]

# # Calcular o ponto central
# latitudes = [point[0] for point in data]
# longitudes = [point[1] for point in data]
centro_lat = sum(latitudes) / len(latitudes)
centro_lon = sum(longitudes) / len(longitudes)

# Criar o mapa
mapa = folium.Map(location=[centro_lat, centro_lon], zoom_start=14)

# Adicionar o HeatMap
HeatMap(data[:1000], radius=15, max_zoom=13).add_to(mapa)
mapa
# Salvar o mapa
# mapa.save('heatmap_ficticio.html')
```

```python

# Criar um mapa centralizado nas coordenadas médias
m = folium.Map(location=[df['geolocation_lat'].mean(), df['geolocation_lng'].mean()], zoom_start=13)

# Converter as colunas de latitude e longitude em uma lista de listas
heat_data = [[row['geolocation_lat'], row['geolocation_lng']] for index, row in df.iterrows()]

# Adicionar o heatmap ao mapa
HeatMap(heat_data).add_to(m)

# Salvar o mapa como um arquivo HTML
# m.save("heatmap_from_dataframe.html")
m
```
