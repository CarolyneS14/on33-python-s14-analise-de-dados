- JOIN

```python
#Merge para unir o df_pedidos e o df_itens_pedido
df_olist_pedidos = pd.merge(df_pedidos, df_itens_pedido, on=['order_id'], how='inner')
df_olist_pedidos
```

```python
# Verificar as colunas depois do JOIN
df_olist_pedidos[df_olist_pedidos.price.isna() == True]
```

```python
#Merge para unir o novo df_olist_pedidos e o df_consumidor
df_olist_pedidos_uf = pd.merge(df_olist_pedidos, df_consumidor, on=['customer_id'], how='inner')
df_olist_pedidos_uf
```

```python
# Verificar as colunas depois do JOIN
df_olist_pedidos_uf[df_olist_pedidos_uf.price.isna() == True]
```

```python
#Merge para unir o novo df_olist_pedidos_uf e o df_produtos
df_olist_pedidos_uf_prod = pd.merge(df_olist_pedidos_uf, df_produtos, on=['product_id'], how='inner')
df_olist_pedidos_uf_prod
```

```python
# Verificar as colunas depois do JOIN
df_olist_pedidos_uf_prod[df_olist_pedidos_uf_prod.price.isna() == True]
```

```python
#Merge para unir o novo df_olist_pedidos_uf_prod e o df_produtos
df_vendas_olist = pd.merge(df_olist_pedidos_uf_prod, df_vendedor, on=['seller_id'], how='inner')
df_vendas_olist
```

### O número de vendas varia de acordo com a época do ano?

```python
#Tabela de Sazonalidade ao londo dos anos
df_sazonalidade = df_vendas_olist.groupby('reference_month').agg({'order_id': 'nunique'})
df_sazonalidade.sort_values(by='order_id', ascending=False)
```

```python
#Gráficos de linha analise ao longo dos anos
plt.figure(figsize=(15,5))
sns.lineplot(data=df_sazonalidade.reset_index().sort_values("reference_month"), x="reference_month", y="order_id")
plt.xticks(rotation=45);
```

Não parece existir um padrão de vendas ao longo dos anos e a tendência de crescimento da curva, pode dificultar a visualização de uma sazonalidade mensal.

- Vamos analizar as vendas por dia.

```python
# Sazonalidade por dia
date_sazonality = df_vendas_olist.groupby('reference_date').agg({'order_id': 'nunique'}).reset_index().sort_values("reference_date")
```

```python
#Gráficos de linha analise por dia
plt.figure(figsize=(15,5))
sns.lineplot(data=date_sazonality, 
             x="reference_date", y="order_id")

# Limitar o número de ticks no eixo x
plt.xticks(date_sazonality['reference_date'][::10], rotation=45)  # Altere o valor de '5' para outro intervalo desejado
plt.show()
```

```python
#Gráficos de linha analise por dia de JAN a MAR de 2018

# Convertendo as datas para objetos datetime.date
data_inicio = pd.to_datetime("2018-01-01").date()
data_fim = pd.to_datetime("2018-03-01").date()

# Filtrando os dados e criando o gráfico de linha
plt.figure(figsize=(15, 5))
sns.lineplot(data=date_sazonality[(date_sazonality.reference_date >= data_inicio) & (date_sazonality.reference_date <= data_fim)], 
             x="reference_date", y="order_id")
plt.xticks(rotation=45)
plt.show()
```

### O tempo limite de entrega das compras está sendo respeitado?

```python
#Transformando as colunas em datetima
df_vendas_olist['order_delivered_customer_date'] = pd.to_datetime(df_vendas_olist['order_delivered_customer_date'])
df_vendas_olist['order_delivered_date'] = df_vendas_olist['order_delivered_customer_date'].dt.date
df_vendas_olist['shipping_limit_dt'] = pd.to_datetime(df_vendas_olist['shipping_limit_date']).dt.date

# Classificando
df_vendas_olist.loc[df_vendas_olist['order_delivered_date'] > df_vendas_olist['shipping_limit_dt'], 'status_entrega'] = 'ATRASADO'
df_vendas_olist.loc[df_vendas_olist['order_delivered_date'] < df_vendas_olist['shipping_limit_dt'], 'status_entrega'] = 'ADIANTADO'
df_vendas_olist.loc[df_vendas_olist['order_delivered_date'] == df_vendas_olist['shipping_limit_dt'], 'status_entrega'] = 'DENTRO DO ESPERADO'
df_vendas_olist
```

```python
# Porcentagem de pedidos com base na classificacao
df_vendas_olist["status_entrega"].value_counts(normalize=True)*100
```

Cerca de 72,76% dos pedidos sao entregues com atraso e cerca de 18.32% sao entregues adiantados, e somente 8,9% dos pedidos sao entregues dentro do prazo.
